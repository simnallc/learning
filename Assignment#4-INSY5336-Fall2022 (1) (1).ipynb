<b>Q#1:</b> Consider the Contents.txt file. Using regular expressions,write a program to find: 
<ul>
    <li>Full email addresses.</li> 
    <li>words with exactly 7 letters using regex.</li>
    <li>The numbers starting with 212.</li>
    <li>The html tags that are more than 4 letters. Html tags can be found inside <> characters and closing html tags can be found in the same format after / character. </> </li>
<li>You are given stock prices for related financial tickers. (Symbols representing companies in the stock market).Find the tickers mentioned in the report. i.e.: TSLA, NFLX </li>

```python
import re
f = open("Contents.txt", "r")
file_content = f.read() # read content from file
match = re.findall(r'[\w.+-]+@[\w-]+\.[\w.-]+', file_content)
print("All emails: "+str(match)) # print all emails list

#get all seven letters words list
seven_letters_word = re.findall(r'\b\w{7}\b', file_content)
print("Seven words letters: "+str(seven_letters_word))

#get all numbers starting with 212
print("Numbers starting with 212: ")
all_digits_list = re.findall(r'\d+', file_content)
for digits in all_digits_list:
    if re.match(r'^212', digits):
        print(digits)

#get all html tag text length > 4
print("all html tags text with length > 4: ")
data = re.findall(r'>(.+?)<', file_content)
for d in data:
   if len(d) > 4 and ">" not in d:
       print(d)
#get All tickers 
# Find all tickets
all_tickers = re.findall(r'(?<=\$)\w+|[A-Z]{3,6}', file_content)
print("All tickers list: "+str(all_tickers))
```
